```python
#Reading file
df = (pd.read_csv('317233692_T_ONTIME.csv')
        .rename(columns=str.lower))
```

```python
#ix
df.head()
first = df.groupby('airline_id')[['fl_date', 'unique_carrier']].first()
first.head()
```

```python
#Because index is integer based ix will use it as label based and it don't find 10:15 labels in the indexes
first.ix[10:15,:]
```

```python
#Because now second has label indexes ix will be position based it will print unique_carriers for 10:15 th position
second = df.groupby('unique_carrier').first()
second.ix[10:15,:]
```

```python
#Label based indexing .loc
second.loc[['AA', 'AS', 'DL'], ['fl_date', 'tail_num']]
```

```python
#Position based indexing
second.iloc[[0, 1, 3], [0, 1]]
```

```python
#For this data frame take the rows of b where a was 3 or less, and set them equal to b / 10
f = pd.DataFrame({'a':[1,2,3,4,5], 'b':[10,20,30,40,50]})
f
```

```python
#Practise of multiple indexing f[f['a']<=3]['b']must be avoided in pandas because for _getitem (f[f['a']<=3]) we don't know whether it returns view or copy
#Correct way is using .loc and avoiding f[][] multiple indexing
f.loc[f['a']<=3,'b'] = f.loc[f['a']<=3,'b']/10
f
```

```python
#MultiIndexing using (.set_index())
#Levels of multiIndex are unique_carrier, origin, dest, tail_num, fl_date.. position 0,1,2,3,..
hdf = df.set_index(['unique_carrier', 'origin', 'dest', 'tail_num', 'fl_date']).sort_index()
hdf.head()
```

```python
#Slicing multiIndex
#Outermost Indexing can be simple .loc
#for columns dep_time and dep_delay where the carrier was American Airlines, Delta, or US Airways
hdf.loc[['AA','DL','US'],['dep_time', 'dep_delay']]
```

```python
#for specific second index use touples
hdf.loc[(['AA', 'DL', 'US'], ['ORD', 'DSM']), ['dep_time', 'dep_delay']]
```

```python
#pd.IndexSlice
hdf.loc[pd.IndexSlice[:,['ORD','DSM']],['dep_time','dep_delay']]
```

```python
#similarly for third and fourth index
hdf.loc[pd.IndexSlice[:, 'ORD', 'DSM', :, '2014-01-01':'2014-01-15'],
        ['dep_time', 'dep_delay', 'arr_time', 'arr_delay']]
```
